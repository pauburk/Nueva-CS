```python
# imports

import sklearn
import pandas
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
# import data into dataframe

df = pandas.read_csv('./happiness-report-data/2019.csv') # , sep=';'
df.head()
```

```python
# make a list with the regions and ranks and then remove that column from the df to standardize the data
# If I was actually creating models and using this data, I wouldn't standardize it, this is just to practice

# I also couldn't figure out a way to run through the rows for "Country or region" and replace those with sequential numbers, so I ended up using a for loop...
# And yes I realize that both the "Overall rank" and indecies have the sequential numbering I want, but I still want to figure out how to do that

regions = []
ranks = []

for index, row in df.iterrows():
    regions.append(row["Country or region"])
    ranks.append(row["Overall rank"])
    
df = df.drop("Country or region", axis=1)
df = df.drop("Overall rank", axis=1)
df.head()


# i = -1
# def convert_sequential(row):
#     regions.append(row)
#     i += 1
#     return i
    
# converted_df = df['Country or region'].apply(convert_sequential)
# print(type(converted_df))
# print(converted_df)
# print(regions)

# enc_mjob = preprocessing.LabelEncoder()
# encoded_mjob = enc_mjob.fit_transform(df['Country or region'])
# print(encoded_mjob)
```

```python
# standardize the data even though we don't need to (just to practice)
# also add back in the "Country or region" and "Overall rank" columns

scaler = StandardScaler()
df_s = pandas.DataFrame(data=scaler.fit_transform(df), columns=df.columns)
df_s.insert(0, "Overall rank", ranks) # insert column where it was before (not just adding it to the end)
df_s.insert(1, "Country or region", regions) # insert column where it was before (not just adding it to the end)
df_s.head()
```
